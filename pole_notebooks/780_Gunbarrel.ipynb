## Gunbarrel LIP ca. 780 Ma paleomagnetic pole

### Geologic context

The ca. 780 Ma Gunbarrel large igneous province led to magmatic activity on the western margin of Laurentia that is expressed on the margin of Wyoming craton within the Mackenzie Mountains and on the margin of the Slave Craton. Gunbarrel magmatism has been attributed to upwelling asthenosphere (potentially associated with a mantle plume) that initiated the breakup of the supercontinent Rodinia.

Ding et al. (2025) present new paleomagnetic data from nine ca. 780 Ma mafic dikes in the Beartooth Mountains of the Wyoming craton and one mafic sheet (Calder) from the Slave craton. The Wyoming dikes are fine- to medium-grained diabase (clinopyroxene, orthopyroxene, plagioclase, minor biotite, hornblende, and interstitial granophyre) trending west-east to WNW-ESE with widths of 1–20 m. 

### Paleomagnetic and rock magnetic data

Thermal demagnetization data and thermomagnetic susceptibility experiments are consistent with remanence being dominantly held by low-titanum magnetic.

All samples were subjected to stepwise thermal demagnetization; sister samples from four sites (BT51, BT5658, BT69, J20B24) were also AF demagnetized. High-temperature components (HTC) were isolated between 500–580°C, directed west with shallow inclinations.

The primary nature of the HTC is supported by two positive baked-contact tests: (1) the Christmas Lake dike (J20B24) cross-cutting an older NW-SE dike in the Beartooth Mountains, and (2) the Calder sheet (R10HL8) at Hottah Lake where host rock directions change with distance from the intrusion.

The revised pole combines 11 VGPs from the Wyoming craton (9 new Ding et al. sites plus BT-8 and MM1 from Harlan et al., 1997) with 3 VGPs from the Slave craton (Gunbarrel, Margaret, and Calder sheets; Park, Buchan & Gandhi, 1995; Harlan et al., 2003) to yield a mean pole at 3.2°N, 151.5°E (A95 = 8.0°, N = 14). This pole passes the Deenen et al. (2011) criteria (A95_min = 4.2°, A95_max = 15.6°) and fulfills six of the seven Meert et al. (2020) quality criteria. Ding et al. (2025) argue that previously reported ~780 Ma poles from the Tobacco Root Mountains are biased by a local ~27° clockwise rotation and should not be used for Laurentia reconstructions.

### Age constraints

- BT-8 (Christmas Lake dike): 779.5 ± 3 Ma (U-Pb ID-TIMS baddeleyite; Harlan et al., 2003); 774 ± 4 Ma (⁴⁰Ar/³⁹Ar; Harlan et al., 1997)
- MM1 (Mount Moran dike): 775 ± 10 Ma (⁴⁰Ar/³⁹Ar; Harlan et al., 1997)
- Gunbarrel sill: 779.6 ± 1.4 Ma (U-Pb ID-TIMS baddeleyite; Harlan et al., 2003)
- Calder sheet: 779.5 ± 1.8 Ma (U-Pb ID-TIMS baddeleyite; Harlan et al., 2003)

# Nordic workshop ranking

Prior Nordic Workshop compilations of paleomagnetic poles with expert-assigned rankings (Evans et al., 2021) have given the pole from the Gunbarrel LIP a 'B' ranking. Our assessment is that with the work of Ding et al. (2025) that the revised pole can now be assigned 'A' ranking. There are two major advances with the work of Ding et al. (2025):
- Two positive baked contact tests published in their study: one from the Christmas Lake dike in the Beartooth Mountains of the Wyoming Craton and the other from the Hottah Lake (Calder) gabbro sheet of the Slave craton.
- Numerous additional VGPs from units that are associated with cratons rather than mobile belts combined with the informed exclusion of data from the Tobacco Root Mountains.

```python
import pandas as pd
import numpy as np
import warnings

def make_nordic_summary(terrane, 
                        rockname, 
                        sites,
                        dir_mean,
                        pole_mean,
                        study_lon,
                        study_lat,
                        component_comment='',
                        tests='',
                        f_factor=1,
                        pole_mean_unflattened=None
                        ):
    nordic_dict = {}
    nordic_dict['terrane'] = terrane
    nordic_dict['rockname'] = rockname
    nordic_dict['component_comment'] = component_comment
    nordic_dict['tests'] = tests
    if sites_tc['dir_tilt_correction'].nunique() != 1:
        warnings.warn(
            "Multiple tilt correction values found in sites data; "
            "cannot determine single tilt correction for summary, choosing first value."
        )    
    nordic_dict['tilt'] = sites['dir_tilt_correction'][0]
    nordic_dict['study_lat'] = study_lat
    nordic_dict['study_lon'] = study_lon
    nordic_dict['site_n'] = pole_mean['n']
    nordic_dict['sample_n'] = sites['dir_n_samples'].sum()
    nordic_dict['dir_dec_mean'] = dir_mean['dec']
    nordic_dict['dir_inc_mean'] = dir_mean['inc']
    nordic_dict['dir_inc_mean_abs'] = np.abs(dir_mean['inc'])
    nordic_dict['dir_k'] = dir_mean['k']
    nordic_dict['dir_alpha_95'] = dir_mean['alpha_95']
    nordic_dict['pole_lat'] = pole_mean['lat']
    nordic_dict['pole_lon'] = pole_mean['lon']
    nordic_dict['pole_dp'] = ''
    nordic_dict['pole_dm'] = ''
    nordic_dict['pole_A95'] = pole_mean['alpha95']
    nordic_dict['f_factor'] = f_factor
    nordic_dict['inc_f'] = ipmag.unsquish(dir_mean['inc'], f_factor)
    if pole_mean_unflattened is not None:
        nordic_dict['pole_lat_unflattened'] = pole_mean_unflattened['lat']
        nordic_dict['pole_lon_unflattened'] = pole_mean_unflattened['lon']
        nordic_dict['pole_dp_unflattened'] = ''
        nordic_dict['pole_dm_unflattened'] = ''
        nordic_dict['pole_A95_unflattened'] = pole_mean_unflattened['alpha95']
    else:
        nordic_dict['pole_lat_unflattened'] = pole_mean['lat']
        nordic_dict['pole_lon_unflattened'] = pole_mean['lon']
        nordic_dict['pole_dp_unflattened'] = ''
        nordic_dict['pole_dm_unflattened'] = ''
        nordic_dict['pole_A95_unflattened'] = pole_mean['alpha95']
    
    
```
