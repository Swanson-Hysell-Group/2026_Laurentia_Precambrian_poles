# Adirondack Highlands ca. 887 Ma paleomagnetic pole

## Geologic context

The Adirondack Highlands of northern New York preserve a granulite-facies metamorphic record from the Grenvillian orogeny. Brown and McEnroe (2012) collected paleomagnetic samples from an east-west traverse across the Highlands, sampling three principal rock units: (1) microcline gneisses (GMS) of the western and far-western Highlands, (2) the metamorphosed anorthosites of the Marcy Massif and associated metamorphic rocks (metasyenites, metadiorites, granitic gneisses), and (3) post-metamorphic fayalite-bearing granites of the Wanakena and Ausable Forks plutons. Granulite-facies metamorphism (T > 650 °C) at ca. 1050 Ma completely reset pre-existing magnetizations, with new remanence acquired during slow post-Grenvillian cooling.

Each of the three rock units carries a stable remanence with characteristic dual polarity. The remanence is carried by ilmeno-hematite in the microcline gneisses and by magnetite in the metamorphosed anorthosites (with subsidiary hemo-ilmenite) and in the post-metamorphic fayalite granites. Brown and McEnroe (2012) reported a pole at 18.4°S, 151.1°E for the gneisses, 25.1°S, 149.0°E for the anorthosite group (combining 11 anorthosite sites with 3 associated metamorphic rock sites for a 14-site grouping in their Table 3), and 28.4°S, 132.7°E for the fayalite granite.

Zhang et al. (in revision, *Tectonics*) develop new U-Pb petrochronology and thermochronology datasets from zircon, garnet, titanite, rutile, and apatite from the Adirondack Highlands. Probabilistic thermal history inversion of these data shows that the Adirondack Highlands cooled slowly throughout the Tonian, with the magnetite-carried remanence in the Marcy Massif anorthosite acquired ca. **887 ± 23 Ma** — about 85 Myr later than the ca. 970 Ma cooling-curve estimate of Brown and McEnroe (2012). Zhang et al. additionally report a positive baked-contact test from a dike (MA11) that crosscuts the anorthosite (likely Ediacaran), demonstrating that the anorthosite remanence predates the Phanerozoic. This revised age fills a critical gap in the Laurentia apparent polar wander path between the ca. 990 Ma Jacobsville Formation pole and the ca. 780 Ma Gunbarrel pole, with implications for the position of Rodinia in the early to mid-Tonian.

## Paleomagnetic and rock magnetic data

Brown and McEnroe (2012) characterized the magnetic mineralogy of the Adirondack rocks using SEM petrography of oxide microstructures, isothermal remanent magnetization (IRM) acquisition, hysteresis loops, low-temperature susceptibility, and thermomagnetic experiments. The microcline gneisses are dominated by ilmeno-hematite with multiple generations of fine ilmenite exsolution lamellae, which give rise to the distinctive lamellar magnetism of these rocks. The metamorphosed anorthosites and the fayalite granites are dominated by magnetite (with subsidiary hemo-ilmenite in some anorthosites). The hemo-ilmenite remanence in the microcline gneisses is interpreted by Brown and McEnroe (2012) as a chemical remanent magnetization (CRM) acquired during sub-solvus exsolution of ilmenite lamellae from a titanohematite host, whereas the magnetite-carried remanence in the metamorphosed anorthosites and fayalite granites is a thermoremanent magnetization (TRM) acquired on cooling through the magnetite Curie temperature. Given current knowledge, it is more straight-forward to time calibrate the magnetite-held TRM.

Site mean directions were determined after progressive alternating-field (AF) and thermal demagnetization, with characteristic directions isolated by principal component analysis (Kirschvink, 1980). Sites with normal polarity (steep-down, SE declination) and reversed polarity (steep-up, NW declination) occur in all three rock units. Brown and McEnroe (2012) note that the metamorphic rocks pass the McFadden and McElhinny (1990) reversal test as Class C (γ_o = 4.88° vs. γ_c = 4.68°, slightly outside the 95% level), while the fayalite granite passes the test as Class C (limited by only one normal-polarity site).

## Three poles, three locations

The MagIC contribution for this study reports three pole results — one per location — following the groupings of Brown and McEnroe (2012) Table 3:

1. **Adirondack microcline gneiss** (14 sites, ilmeno-hematite carrier; 900 ± 100 Ma)
2. **Adirondack metamorphic anorthosite** (11 anorthosite + 3 associated metamorphic rock sites = 14 sites total, magnetite carrier; 887 ± 23 Ma per Zhang et al., in press)
3. **Adirondack fayalite granite** (8 sites, magnetite carrier; 900 ± 100 Ma)

The R-criteria assessment below focuses on the 887 Ma metamorphic anorthosite pole — the result on which the new Zhang et al. age constraint operates.

## Site-level data

All 36 sites are stored at `dir_tilt_correction = 0` (geographic coordinates) since these basement metamorphic and igneous rocks have no bedding tilt to correct. The MagIC `location` field assigns each site to one of the three pole groupings.

```python
pt.plot_site_map(sites_geo, zoom_start=8)
```

## Mean poles by rock unit

Each location's pole is computed as a Fisher mean of site VGPs after polarity unification (`unify_polarity=True`). All three units have a mix of normal (AN) and reverse (AR) polarity sites; the unification flips the minority polarity to common reverse polarity before averaging.

```python
for loc in ['Adirondack microcline gneiss',
            'Adirondack metamorphic anorthosite',
            'Adirondack fayalite granite']:
    sub = sites_geo[sites_geo['location'] == loc].reset_index(drop=True)
    _, pole = pt.compute_mean_pole(sub, unify_polarity=True)
    print(f'\n=== {loc} (N={pole["n"]}) ===')
    ipmag.print_pole_mean(pole)
```
