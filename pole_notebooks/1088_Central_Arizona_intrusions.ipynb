## Central Arizona diabase ca. 1088 Ma paleomagnetic pole (N1 population)

### Geologic context

There are widespread Proterozoic thick diabase intrusions in Central Arizona in southwestern Laurentia. Previous paleomagnetic data from these thick mafic intrusions show at least two distinct populations of paleomagnetic directions---one population with shallow westward directions (with one potential coeval intrusion with a reversed polarity), and the other population with steep N and S dual-polarity directions. 

These intrusions are not well dated and the paleomagnetic data is poorly paired with geochronology data. Harlan, 1993 did a pilot paleomagentic study on the sills and Donadini et al., 2011 reported an updated paleomagnetic study from the intrusions (including resampling some sills from Harlan, 1993 and regrouping paleomagnetic data into a unified site-level dataset). In SUPERCONTINENT SYMPOSIUM 2012 abstract compilation Donadini et al reported that one dike with the N1 direction (shallow and westerly reported in Donadini et al., 2011) is dated at 1088 ± 11 Ma (2 $\sigma$) by U-Pb geochronology. However, that date comes from personal communication with Ryan Bright and the data was not presented with mineral source or a data table and the field guide is not peer-reviewed. 

In context of the Keweenawan Track paleomagnetic apparent polar wander path, the two populations of the paleomagnetic directions from the intrusions are possibly associated with an early pulse coincident with the early stage magmatism of the Midcontinent Rift before ca. 1105 Ma Alona Bay reversed polarity zone, whereas the population with shallower inclination possibly correlates with the younger pulse of magmatism in the MCR ca. 1096 Ma but could also be associated with the Portage Lake Volcanism. In Mohr et al., 2024, one mafic sill from the Salt River Canyon in Central Arizona was dated at 1097.97 ± 0.12 Ma which is consistent with it being associated with the mafic magmatism of the SWLLIP at the time. However, it is uncertain whether that dated sill was previously sampled for paleomagnetic study of Donadini et al., 2011.

### Paleomagnetic and rock magnetic data

Harlan, 1993 conducted both AF and thermal demagnetization on their samples. PCA analysis was used to isolate the characteristic remanent magnetization (ChRM) directions. The demag data and IRM acquisition experimental data were used to infer magnetic mineralogy. Authors interpreted that the remanence is dominantly held by low-titanium magnetite.

Donadini et al., 2011 conducted both AF and thermal demagnetization on their samples. Additional low-temperature demagentization was carried out for some of the samples.PCA analysis was used to isolate the characteristic remanent magnetization (ChRM) directions. In addition, the study measured hysteresis loops, x-T experiments, and petrographic and SEM data from their samples. Titanomagnetite and titanohematite were identified as the main magnetic carriers in the samples. 

Donadini et al., 2011 conducted a baked contact test on one sill (ED) that yielded a shallow and westerly direction (N1 population). The baked quartzite sedimentary rock samples of the Pioneer near the sill yielded a direction that overlaps with the sill direction. That direction is close to but distinct from previously reported paleomagnetic directions from the Pioneer quartzite in Scott and Paulo, 1976. The baked contact test results suggest that the shallow and westerly direction (N1 population) is likely a primary magnetization component. 

The study did not report tilt measurement nor did it apply any tilt correction to the paleomagnetic data. However, the authors did adopt the correction of Harlan (1993) for a consistent rotation of 5° associated with the hypothesized Colorado Plateau rotation.

The site groupings of the sampled sills Zhang et al., 2024 re-compiled the paleomagnetic data from both studies and re-evaluated the groupings. A proposed site-level mean pole is calculated for the N1 population which we will reproduce here.

## Harlan (1993) Diabase sheets (sills) (Central Arizona)

The sample sites are adjacent and within similar "cooling" units. We compile data from this study and attempt to combine the pmag results with those from the study of Donadini et al., 2011 study and try to distinguish individual cooling units and calculate the mean poles for the normal polarity sites and the reversed polarity sites. The main issue that we had to deal with was that the two studies had ambiguous overlap in sampled sills. 

According to the authors, demagnetization methods used in experiments are progressive AF or thermal demagnetizations. Least square algorithm was used in doing line fits.

Issues related to this study and Donadini (2011) study in terms of overlapping sites and tilt correction, and rotation correction based on the Colodado Plateau:
- Harlan's sites are all named as "GD##", which is different from the single site named "GD" in Donadini (2011) study. 
- The site GD in Donadini (2011) does not show up on their geologic map, because the site locations is exactly the same as site DC. 
- However, the samples used for paleointensity study and reported unreliable results in Donadini (2011) study are from the GD sites from Harlan (1993). 
- Harlan (1993) reported tilt correction for two sites, GD25 and GD26, which are further south from other sites. Donadidni (2011) mentioned that the amount of tilt in the region is small and thuse no tilt correction was applied. This is likely related to that the Harlan study is that they did not follow as strict field rules as Helsley to only sample diabase sites with good paleohorizontal constraint. 
- Donadini (2011) has inconsistent descriptions about rotation correction in their table 5 and subsequent texts. We are not sure if the data reported in table 5 is actually corrected for 5 $^\circ$ rotation.
- They attribute the unusual scatter in their data to be due to a combination of poor sampling, not being unable to isolate the primary components, and Laramide orogeny remagnetizations.

We are not including site GD21 because it is likely remagnetized; we are not including SR sites in Harlan (1993) because they are from another study by Sandberg and Butler (1986). Sites GD18, GD19, GD20 are interpreted to be significantly affected by Laramide Orogeny and were excluded by Harlan et al. 

Reference:
Harlan, S. S. (1993). Paleomagnetism of Middle Proterozoic diabase sheets from central Arizona. Canadian Journal of Earth Sciences, 30(7), 1415-1426. DOI: 10.1139/e93-122.

## Donadini (2011) Diabase sheets (sills) (central Arizona)

Samples from 33 sites of diabase sills were collected from central Arizona. Step-wise thermal and AF treatments were both applied to isolate characteristic magnetizations. Principal component analysis were performed using the PmagPy software. Two distinct magnetic poles were interpreted. They did not correct for the rotation of the Colorado plateau (5 degree). 

Some sites were excluded from paleodirection anlysis based on rock magnetic studies: DF, ND, SD, VD, WD, XD3. Their final paleomagnetic analysis were based on selection criteria: MAD < 15, site a95 < 15, and k > 30, and N > 3. As a result the resultant mean pole position for the reversed pole has large uncertainty. 

Donadini, F., Pesonen, L.J., Korhonen, K., Deutsch, A. and Harlan, S.S., 2011. Paleomagnetism and paleointensity of the 1.1 Ga old diabase sheets from central Arizona. Geophysica, 47(1-2), pp.3-30.

```python
Donadini2011a_sites = ipmag.vgp_calc(Donadini2011a_sites,tilt_correction='no',site_lon='lon',site_lat='lat',dec_is='dir_dec',inc_is='dir_inc')
# Dropping ND, SD, VD, XD3 according to authors' rejection
Donadini2011a_sites = Donadini2011a_sites.drop(Donadini2011a_sites[Donadini2011a_sites['site']=='ND'].index)
Donadini2011a_sites = Donadini2011a_sites.drop(Donadini2011a_sites[Donadini2011a_sites['site']=='VD'].index)
Donadini2011a_sites = Donadini2011a_sites.drop(Donadini2011a_sites[Donadini2011a_sites['site']=='XD3'].index)

Donadini2011a_sites
```

```python
ipmag.plot_net()
ipmag.plot_di(Donadini2011a_sites.dir_dec.tolist(),Donadini2011a_sites.dir_inc.tolist())
```

## merging data from the two studies and resolve overlapping sites between Harlan (1993) and Donadini (2011)

1. We will primarily follow Harlan (1993) because they split the sills into individual cooling units.

2. GD27 == GD15 (the question is, are ID, DE, and WD from same sill as these, and are them three from same sill? in Arizona state map they are not!) Looks like the GD15 and GD27 are the same as DE. not sure if the ID and WD are the same as DE sill.
So we will not take DE

3. Donadini DC and GD looks to be replicates of Harlan 29, 30. But Harlan splits two sills between 29 and 30. 
So we will not take DC or GD in Donadini. 

4. The Sierra Ancha sites from Donadini are unique but we don't know whether they are from same sills. 

5. Harlan GD12 and GD13 are the same sill which was also sampled as OD in Donadini, so we will calculate a mean direction for the sill and use the new mean as the site level data. 

6. We interpret that the Donadini KD and LD are the same sill, due to close proximity and anomolous low inclination

7. for the salt river region data, we preferentially use Harlan data because they are split into sills. So we should drop all Donadini stes in Salt River region. 

8. we should drop the Harlan GD23 site because it is anomolous inclination of 0. 

9. Donadini BD and CD are likely the same sill, and be included since Harlan does not have reversed sill from Salt lake river area. BD has better kappa parameter so we choose to keep site BD. 

10. KD and LD are likely from the same sill. But site LD has much larger uncertainty.  

11. WD and ID in Donadini are close and have similar directions. They might be from the same sill. We use WD here because it has more samples and have tighter uncertainty. 

12. site XD3 according to Donadini is not diabase and they dropped it. 

13. VD was also dropped by Donadini. 

14. SD appears to be a single site where an antipodal component to the normal polarity component was found. There is not enough data to give further insight into the origin of that direction. Thus we exclude it from our analysis.

15. We drop Harlan GD23 site because the anomolously 0 inclination angle.

In summary, the steeeply reversed directions in central Arizona are from GD11, GD15-27, and BD, WD. 

```python
# filter for the sites to keep in Donadini data.
Donadini2011a_sites = Donadini2011a_sites[Donadini2011a_sites['site'].isin(['DF', 'DG', 'DJ', 'MD', 'BD', 'CD', 
                                                                            'KD', 'LD', 'ID', 'WD', 'SD', 'OD'])]
Harlan_Donadini = pd.concat([Harlan1993a_sites, Donadini2011a_sites], ignore_index=True)

# keep BD instead of CD for better uncertainty (they are same sill)
# keep WD instead of ID for better uncertainty (they are same sill)
# drop Donadini LD site and keep KD site because of close proximity and similar low inclination but LD has much larger uncertainty 
Harlan_Donadini = Harlan_Donadini.drop(Harlan_Donadini[Harlan_Donadini['site'] == 'LD'].index)
Harlan_Donadini = Harlan_Donadini.drop(Harlan_Donadini[Harlan_Donadini['site'] == 'CD'].index)
Harlan_Donadini = Harlan_Donadini.drop(Harlan_Donadini[Harlan_Donadini['site'] == 'ID'].index)

# drop Harlan GD23 site because the anomolously 0 inclination angle
Harlan_Donadini = Harlan_Donadini.drop(Harlan_Donadini[Harlan_Donadini['site'] == 'GD23'].index)

# take average of GD12, GD13 and OD as they are likely one same sill, the individual entry will be dropped and we calculate a mean direction from the three records
GD12_GD13_OD = Harlan_Donadini[Harlan_Donadini['site'].isin(['GD12', 'GD13', 'OD'])]
GD12_GD13_OD_mean = ipmag.fisher_mean(GD12_GD13_OD['dir_dec'].tolist(), GD12_GD13_OD['dir_inc'].tolist())
Harlan_Donadini = pd.concat([Harlan_Donadini, pd.DataFrame([{'site':'GD12_GD13_OD',
                                          'location': 'Gila County',
                                          'dir_dec': GD12_GD13_OD_mean['dec'],
                                          'dir_inc': GD12_GD13_OD_mean['inc'],
                                          'dir_k': GD12_GD13_OD_mean['k'], 
                                          'dir_alpha95': GD12_GD13_OD_mean['alpha95'],
                                          'lat': 33.55, 
                                          'lon': -110.69, 
                                          'geologic_classes': 'Intrusive:Igneous', 
                                          'geologic_types': 'Sill', 
                                          'lithologies': 'Diabase',
                                          'method_codes': 'LP-DIR-AF:LP-DIR-T:DE-BFL;LP-DC4',
                                          'result_quality': 'g',
                                          'result_type': 'i', 
                                          'age': '1100', 
                                          'age_unit': 'Ma', 
                                          'citations': 'Donadini2011a'}])])

Harlan_Donadini = ipmag.vgp_calc(Harlan_Donadini,tilt_correction='no',site_lon='lon',site_lat='lat',dec_is='dir_dec',inc_is='dir_inc')

individual_sills = ['GD01-02', 'GD03', 'GD04', 'GD05', 'GD06-07', 'GD08', 'GD09', 'GD10', 'GD12_GD13_OD', 
                'GD17', 'GD18-20', 'GD22', 'GD24', 'GD29', 'GD30', 'GD11', 'GD15-27', 
                       'DF', 'DG', 'DJ', 'MD', 'BD', 'KD', 'WD']
Harlan_Donadini = Harlan_Donadini[Harlan_Donadini['site'].isin(individual_sills)]
Harlan_Donadini = Harlan_Donadini.reset_index(drop=True)

Harlan_Donadini_dir_normal, Harlan_Donadini_dir_steep_reversed = pmag.flip(Harlan_Donadini[['dir_dec', 'dir_inc']].to_numpy())
Harlan_Donadini_pole_normal, Harlan_Donadini_pole_steep_reversed = pmag.flip(Harlan_Donadini[['vgp_lon','vgp_lat']].to_numpy())
```

```python
reversed_sites = Harlan_Donadini[Harlan_Donadini['dir_inc'] < 0]['site'].tolist()
Harlan_Donadini_normal_vgp_df = Harlan_Donadini[~Harlan_Donadini['site'].isin(reversed_sites)].reset_index(drop=True)
Harlan_Donadini_normal_vgp_df
```
